```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dis_com = pd.read_excel(r'C:\Users\Lenovo\1代码_PSG课题\代码上传\data\figure 4a_disease.xlsx',
                       index_col = [0])
stage_com = pd.read_excel(r'C:\Users\Lenovo\1代码_PSG课题\代码上传\data\figure 4a_stage.xlsx',
                       index_col = [0])
```

```python
sns.set_context("paper", font_scale=1.5)
fig, ax = plt.subplots(figsize=(3 ,4))
box_plot = sns.boxplot(x='disease' , y='acc', data=dis_com, linewidth=2, saturation =1,
                         color = 'skyblue',  width = 0.4,
                       boxprops=dict(alpha=.3))
sns.stripplot(x='disease' , y='acc', data=dis_com,  palette=['red', 'blue', 'orange'],size = 7,alpha=.7)
box_plot.set(ylim = (0.75,1.05), yticks = [  0.8, 0.9, 1.0])
box_plot.set_xticklabels ([ 'dys', 'HD', 'PD'])
box_plot.set(xlabel='', ylabel= 'Decoding accuracy',)
plt.tight_layout()
```

```python
sns.set_context("paper", font_scale=1.5)
fig, ax = plt.subplots(figsize=(3 ,4))
box_plot = sns.boxplot(x='lable' , y='acc', data=stage_com, linewidth=2, saturation =1,
                         color = 'skyblue',  width = 0.4,
                       boxprops=dict(alpha=.3))
sns.stripplot(x='lable' , y='acc', data=stage_com,  palette=['red', 'blue', 'orange'],size = 7,alpha=.7)

box_plot.set(ylim = (0.78,1.05), yticks = [  0.8, 0.9, 1.0])

box_plot.set(xlabel='', ylabel= 'Decoding accuracy',)
plt.tight_layout()
```

```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats, signal
from scipy.stats import zscore
```

```python
wake = pd.read_excel(r'C:\Users\Lenovo\1代码_PSG课题\代码上传\data\figure 2beh_wake.xlsx')
nrem = pd.read_excel(r'C:\Users\Lenovo\1代码_PSG课题\代码上传\data\figure 2beh_nrem.xlsx')
rem = pd.read_excel(r'C:\Users\Lenovo\1代码_PSG课题\代码上传\data\figure 2beh_rem.xlsx')
```

```python
## Awake
fig, ax = plt.subplots(figsize=(4,7))
sns.pointplot(x='power_scaled', y='subj_band', data=wake, hue = 'band', 
              join=False,  ci = 68, scale = 1.05, errwidth = 0.75,
              markers = ['o', '^', '|', 'd', 's', '1',],color ='red',
              ax=ax)
plt.xlabel('Power (z-score)', fontsize=16)
plt.ylabel('Frequency bands*subjects', fontsize=16)
plt.axvline(x=0, ymin=0, ymax=1, linestyle = '--', alpha = 0.8, color = 'gray')
ax.set(yticklabels=[])
fig.tight_layout()
```

```python
## NREM
fig, ax = plt.subplots(figsize=(4,7))
sns.pointplot(x='power_scaled', y='subj_band', data=nrem, hue = 'band', 
               join=False,  ci = 68, scale = 1.05, errwidth = 0.75,
              markers = ['o', '^', '|', 'd', 's','1',],color ='blue',
              ax=ax)

plt.xlabel('Power (z-score)', fontsize=16)
plt.ylabel('Frequency bands*subjects', fontsize=16)
plt.axvline(x=0, ymin=0, ymax=1, linestyle = '--', alpha = 0.8, color = 'gray')
ax.set(yticklabels=[])
fig.tight_layout()
```

```python
## REM
fig, ax = plt.subplots(figsize=(4,7))
sns.pointplot(x='power_scaled', y='subj_band', data=rem, hue = 'band', 
              join=False,  ci = 68, scale = 1.05, errwidth = 0.75,
              markers = ['o', '^', '|', 'd', 's','1',],color ='orange',
              ax=ax)
plt.xlabel('Power (z-score)', fontsize=16)
plt.ylabel('Frequency bands*subjects', fontsize=16)
plt.axvline(x=0, ymin=0, ymax=1, linestyle = '--', alpha = 0.5, color = 'gray')
ax.set(yticklabels=[])
fig.tight_layout()
```
